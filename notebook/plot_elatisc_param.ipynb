```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

coef_dict = {
    "dt_min": 0.0785, "ELEVATION": 0.8410, "DEW": 9.7039, "VIS": 1.9670,
    "CIG": 0.4906, "SLP": 0.3628, "WND_Speed": 3.0288, "TMP_lag1": 82.2718,
    "time_speed": 116.7931, "time_a": 26.7198, "WND_sin": -0.0076,
    "WND_cos": -0.1803, "month_sin": -0.2883, "month_cos": -0.5375,
    "hour_sin": -0.7323, "hour_cos": -0.8161, "lat_sin": 0.3000,
    "lat_cos": 1.6825, "lon_sin": 0.0405, "lon_cos": -0.0479
}

coef = pd.Series(coef_dict).sort_values(key=lambda x: abs(x), ascending=False)

figures_dir = Path("figures")
figures_dir.mkdir(exist_ok=True)
out_path = figures_dir / "elastic_feature_contributions.png"

plt.figure(figsize=(12, 6))
plt.bar(coef.index, coef.values, color="tab:blue")
plt.xticks(rotation=45, ha="right", fontsize=10)  # 与 RF 图一致
plt.ylabel("Coefficient Value (scaled feature space)")
plt.title("Elastic Net: Feature Contributions")
plt.tight_layout()
plt.savefig(out_path, dpi=300, bbox_inches="tight")
plt.close()
print(f"✅ Saved figure to {out_path.resolve()}")
```
